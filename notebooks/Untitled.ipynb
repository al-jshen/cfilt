```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import gaussian_filter
from scipy.signal import medfilt, savgol_filter
```

```python
x = np.linspace(0, 10)
y = np.sin(x)
yn = y + np.random.normal(size=x.size)
plt.plot(x, yn)
plt.plot(x, y)
```

```python
z = np.convolve(yn, 1 / 4 * np.array([1, 2, 1]), mode="same")
z2 = np.convolve(z, np.array([-1 / 6, 1, -1 / 6]) / (1 + -2 * 1 / 6), mode="same")
# z3 = np.convolve(yn, np.array([-1, 4, 10, 4, -1]) / 16, mode="same")
plt.plot(x, y)
plt.plot(x, yn)
# plt.plot(x, z)
plt.plot(x, z)
plt.plot(x, savgol_filter(yn, 21, 3))
plt.plot(x, medfilt(yn, 5))
plt.plot(x, gaussian_filter(yn, sigma=3), lw=5)
# plt.plot(x, z3, ls="--")
```
